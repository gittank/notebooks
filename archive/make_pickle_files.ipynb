```python
def pickleEventFiles(dbase, run, logFile, tableList, outpath):
    cursor = connectDbase('rct05db')
    log = pd.read_pickle(logFile)
    log = log[log['run']==run]
            
    for eventId in log.eventId.unique():
        s = str(log[(log.eventId==eventId) & (log.state =='1')].date.values[0])    
        e = str(log[(log.eventId==eventId) & (log.state =='0')].date.values[0])    
        for tbName in tableList:
            
            filename = dbase + '_' + run + '_' + str(eventId) + '_' + tbName + '.pkl'
            print 'Pickling %s'%filename
            dfTable = readDateTable(cursor, tbName, s, e)             
            dfTable.to_pickle(outpath + '/' + filename)
```

```python
pickleEventFiles('rct005', '002','Data/rct005/rct005_log.pkl', ['002_highsamp'], 'Data/archive/rct005/data')
```

```python
pickleEventFiles('rct005', '002','Data/rct005/rct005_log.pkl', ['002_lowsamp'], 'Data/archive/rct005/data') 
```
